## Research Objective

The Manning River catchment, located in New South Wales, Australia, is sensitive to changes in vegetation cover due to its varied topography and strong rainfall–runoff dynamics. Vegetation plays a key role in regulating hydrological processes such as interception, infiltration, and evapotranspiration. Disturbances to this vegetation, such as large-scale bushfires, can significantly alter these processes and potentially increase runoff and peak discharge.

The 2019–2020 Australian east coast bushfires were unprecedented in scale, with millions of hectares of temperate forest burned across eastern Australia . In fact, approximately 21% of the temperate broadleaf forest biome burned during a single fire season, far exceeding historical levels . These extreme fires were driven by severe drought conditions and exceptionally low fuel moisture content, which allowed fires to spread across normally fire-resistant landscape features (Boer et al., 2020). The river originates in the Barrington Tops and surrounding high‑elevation forests, where steep, rugged terrain and cool, wet conditions dominate the headwaters. From these upland catchments, the Manning flows east and southeast through a mosaic of forested slopes, agricultural valleys, and rural townships, gradually widening as it approaches the coast. The river passes through the regional centers of Wingham and Taree, supporting irrigation, fisheries, and local communities, before dividing into two tidal channels and entering the ocean near Harrington and Old Bar.

## Catchment Description

The Manning River catchment is located in New South Wales, Australia, draining towards the east coast into the Tasman Sea. The river originates in the elevated regions of the Great Dividing Range and flows eastward through mountainous forested areas, agricultural valleys, and floodplains before reaching the coast near the town of Taree.

The catchment is characterized by steep upper catchment areas and a dense drainage network, making it responsive to rainfall and sensitive to changes in vegetation cover. Land use varies from forest-dominated uplands to agricultural and semi-urban areas in the lower basin.

For this study, the analysis focuses on the sub-catchment corresponding to the Killawarra gauging station, which represents the upstream contributing area used for discharge measurements.

The catchment lies approximately between -31.5°S to -32.5°S latitude and 151.5°E to 152.5°E longitude.
