## Research Objective

The Manning River catchment, located in New South Wales, represents a hydrological system in which vegetation plays an important role in regulating water flows. Vegetation influences processes such as interception, infiltration, soil moisture storage, and evapotranspiration, which together determine how rainfall is divided between runoff and evaporation. Changes in vegetation can therefore directly affect discharge.

A big disturbance to vegetation occurred during the 2019–2020 Australian east coast bushfires, which were unprecedented in both scale and intensity. During this event, hectares of forest were burned across eastern Australia, with approximately 21% of the temperate broadleaf forest biome affected in a single fire season far exceeding historical observations (Boer et al., 2020). These extreme conditions were driven by prolonged drought and very low fuel moisture, allowing fires to spread into areas that are normally less vulnerable to burning. As a result, large parts of catchments like the Manning experienced vegetation loss.

Such vegetation loss can affect hydrological processes by reducing interception and evapotranspiration, and by lowering soil infiltration capacity. This can increase surface runoff and potentially lead to higher peak discharges after rainfall events. Studies have shown that fire-induced vegetation loss can increase streamflow responses, although the strength of this effect depends on catchment characteristics and recovery processes (Nolan et al., 2020).

However, it is still unclear to what extent these changes can be observed in discharge data. Hydrological responses may be influenced by climate variability or other processes.

This leads to the central research question: *To what extent did the 2019–2020 bushfires influence the discharge behavior of the Manning River catchment, and can these effects be identified in observed hydrological data?*




## Catchment Description

The Manning River catchment is located in New South Wales, Australia, and drains eastward into the Tasman Sea. The river originates in the elevated and forested regions of the Barrington Tops within the Great Dividing Range. These headwaters are characterized by steep terrain and relatively high rainfall (NSW, 2021).

From these upland areas, the river flows east and southeast through a varied landscape of forested slopes, agricultural valleys, and rural settlements. Along its course, it passes regional centers such as Wingham and Taree, where it supports irrigation, fisheries, and local communities, before reaching the coast near Harrington and Old Bar.

The catchment has a dense drainage network and steep upper areas, which will impact the rainfall. Land use varies from mainly forest in the upper catchment to agricultural and semi-urban areas in the lower basin.

For this study, the focus is on the sub-catchment corresponding to the Killawarra gauging station with an area of 6642 km$^2$, which represents the upstream area used for discharge measurements. The catchment is located approximately between 31.5°S to 32.5°S latitude and 151.5°E to 152.5°E longitude.

### River Discharge with flow duration curve

The river discharge time series for approximately five years before and five years after the 2019–2020 bushfires were analysed to assess whether a significant change in discharge behaviour can be observed.

![River discharge](Images/discharge_before_after.png)

The discharge time series shows a clear difference between the pre- and post-deforestation periods. Before the 2019–2020 bushfires, discharge values are relatively low and stable, with only occasional peaks. In contrast, the post-deforestation period shows higher variability and more frequent extreme discharge events.

The mean discharge after deforestation is also noticeably higher than before, indicating an overall increase in runoff. This could suggest that vegetation loss has reduced interception and evapotranspiration, leading to a stronger runoff response.

However, the increase in discharge is mainly driven by several extreme peak events rather than a consistent shift in baseline discharge. These peaks are likely related to high precipitation events, which are known to strongly influence discharge.

Therefore, although the results suggest a possible impact of deforestation on discharge dynamics, it is not possible to attribute the observed changes only to vegetation loss. Precipitation variability likely plays a major role in the observed differences.

![FDC](Images/fdc.png)

The flow duration curve indicates higher discharge across all exceedance probabilities after deforestation. This suggests that deforestation increases both peak flows and baseflow.
Flow duration curves are particularly useful as they describe the full distribution of discharge and allow comparison of hydrological behavior under different land-use conditions (Yadav et al., 2007).

### Run-off Coefficient

Because there is a lot of variability in discharge and precip, the monthly means from each year are taken to calculated the runoff coefficient. The run-off coefficient is expected to increase after deforestation, as vegetation plays an important role in intercepting rainfall, increasing infiltration and reducing surface runoff. With reduced forest, less water is intercepted and more precipitation reaches the ground directly. As a result, a larger part of precipitation is expected to be converted into runoff.

![River coefficient](Images/runoff_coefficient.png)

The run-off coefficient shows higher peaks after the deforestation, indicating that more runoff is generated. However, the overall pattern remains really variable and the average values do not show a consistent increase compared to the pre-deforestation. Large fluctuations are already present before deforestation, suggesting that natural variability, particularly in precipitation, influences the run-off coefficient. This makes it difficult to attribute the changes directly to deforestation. Increase in peak runoff can be observed, but the correlation between the increased runoff and deforestation is not consistent. Also, some values exceed 1,
indicating data irregularities that may affect result validity.

### Rootzone Storage

Rootzone storage is expected to decrease after deforestation, as vegetation plays a key role in accessing and retaining soil moisture. With reduced forest cover, the system becomes less capable of storing water and supplying it to plants during dry periods. As a result, rootzone storage deficits are expected to become larger and more persistent, indicating increased drough. Howevetr, ecosystems adapt their root systems to climate and water availability, meaning that rootzone storage is not only controlled by vegetation but also strongly influenced by precipitation patterns and climate conditions (Wang-Erlandsson et al., 2016).

![Rootzone](Images/rootzone_storage_deficit.png)

In the Manning catchment, the rootzone storage deficit shows an increase after the deforestation period, with more negative values and longer-lasting deficits compared to the pre-deforestation period. This suggests that the system experiences higher water stress and reduced recovery during dry conditions. However, the difference is not consistently pronounced throughout the entire time series. Even before deforestation, relatively large deficits are already present, indicating that the catchment naturally experiences substantial variability in soil moisture conditions.

This makes it difficult to attribute the observed changes solely to deforestation. Also, climatic variability, particularly precipitation, plays a big role in controlling rootzone storage. Additionally, uncertainties in the data and assumptions in the water balance approach may influence the magnitude of the deficits. So, while an increase is visible, the overall impact remains moderate and cannot be isolated from other controlling factors.

### Budyko Framework

The budyko graph is used and we checked if deforestation could be detected in it. Because  in theory it leads to less evaporation and increase in runoff, theory says that we should see a shift in the graph (Budyko, 1974).

<img src="Images/budyko_before_after.png" width="500">

The Budyko plot reveals a distinct shift after deforestation, with a decrease of approximately 1 in the aridity index (Ep/P) and about 0.2 in the evaporative index (Ea/P). This indicates that a smaller proportion of precipitation is returned to the atmosphere, while a larger share contributes to runoff or storage. Such a shift is consistent with reduced vegetation cover, which decreases transpiration and interception losses (Brown et al., 2005). According to Budyko (1974), long-term water partitioning is primarily dominated by climate, which explains why the overall change remains relatively small. So, while deforestation has a detectable effect on the water fluxes, climate stays the dominant control on the catchment water balance.

### Su,max of Discharge Models

The parameter Su,max represents the maximum amount of water that can be stored in the root zone and accessed by vegetation. It is strongly influenced by vegetation characteristics, such as root depth, and is therefore expected to decrease after deforestation.
