## Analysis of Deforestation Effects on Catchment Characteristics

To analyse whether the impact of the deforestation in 1995 can be seen in the Cagayan de Oro watershed, several different catchment characteristics were investigated pre- and post-deforestation. The following effects were analysed:
- River discharge (*EFFECTS_ANALYSIS.ipynb*)
- Run-off coefficient (*EFFECTS_ANALYSIS.ipynb*)
- Rootzone storage (*EFFECTS_ANALYSIS.ipynb*)
- Budyko plots (*EFFECTS_ANALYSIS.ipynb*)
- Su,max of discharge models (*CALIBRATION_PRE_DEFORESTATION.ipynb* and *CALIBRATION_POST_DEFORESTATION.ipynb*)

### River Discharge

The river discharge timeseries for 4 years pre- and 4 years post-deforestation were plotted to see whether a significant change can be seen.

Pre-and post-deforestation a significant change in the discharge can be seen of roughly 27%. The discharge data is very variable/peaky, so no significant change in the flashiness of the catchment can be seen.

To assess whether the mean river discharge increase is definitely attributable to the deforestation, several other random samples were tested to see if such a difference occurs more frequent in the discharge series.

It can be seen that big changes in mean river discharge occur at other times in the discharge data as well. Thus, it can be that the increase in mean discharge is because of deforestation, but it is not possible to say this for certain as there are also many other factors (e.g. precipitation) that affect discharge magnitudes. 

### Run-off Coefficient

Another variable that can be influenced by deforestation is the run-off coefficient. Thus, this was plotted to see if any difference could be noticed. As the daily discharge and precipitation variables are very variable, the monthly means were used to calculate the run-off coefficient.

With deforestation an increase in the run-off coefficient would be expected right after deforestation, after which it slowly bounces back to the more original values. Yet, no clear change can be seen in the run-off coefficient during/after deforestation. However, the graph also shows the run-off coefficient is above 1 quite often which should not be possible. Thus, it raises questions about the soundness of the data used for this analysis. 

### Rootzone Storage

The rootzone storage is another characteristic affected by deforestation, which is expected to decrease when forests are cut down.

Contrary to what is expected after deofestation, the maximum rootzone storage increases after deforestation. Apart from this peak, a slight decrease in the annual signal of the rootzone storage can maybe be seen. However, it is difficult to say if this is an actual difference as the catchment already shows a very low rootzone storage before deforestation as well. The behaviour of the rootzone storage suggests that there are other factors, such as precipitation, that play a more important role in the rootzone storage making it hard to see an effect of the deforestation. In addition, earlier questions about the accuracy of the data can also have an effect on the outcome of the rootzone storage making it hard to see and conclude a definite impact.

### Budyko Framework

Deforestation can also be detected in the Budyko Framework, as it leads to a decrease of the evaporation index and increase in the run-off coefficient causing a vertical shift in the graph. 

The Budyko Framework shows a minimal downward shift post-deforestation. This suggests there is little/no effect of deforestation to be seen on the long-term water fluxes of the catchment, which is also in line with the earlier run-off coefficient analysis. This could be because the effect of roughly 10% deforestation is not enough when looking over the entire catchment and/or there are other (climatic) variables that play a more important role in this area.

### Su,max of Discharge Models

The final part of the analysis was to calibrate the discharge model of the Cagayan de Oro catchment to the data before deforestation and after deforestation to see whether a change in the Su,max (maximum unsaturated storage) parameter could be detected. The calibrations for these notebooks were not done in this notebook to prevent long run times, but they can be found in the following files:
- Calibration pre-deforestation: *CALIBRATION_PRE_DEFORESTATION.ipynb*
- Calibration post-deforestation: *CALIBRATION_POST_DEFORESTATION.ipynb*

These files also contain the validation of the models on different time periods.

The results of the calibation based on the De value:
- Su,max pre-deforstation: 767.45 mm 
- Su,max post-deforestation: 54.67 mm 

Su,max for the different parameters sets, according to best NSE, logNSE, De are visualised in the tables below.

Solely looking at the change in maximum unsaturated storage, a significant decrease can be seen which is expected when deforestation occurs, as fewer trees means less rootwater uptake which in turn means less storage in the ground. Thus, this could suggest an effect of deforestation in 1995 can be seen. However, both models do not have a very high NSE, logNSE (and minimal De) value which indicates the model does not fit the observed discharge very well, raising questions about the performance of the model and thus the 'accuracy' of the parameters. Furthermore, when the discharge is modeled for the period post deforestation with the pre-deforestation parameters the model tends to overpredict the discharge most of the time where it would sooner be expected to see un underprediction. 

## Analysis Different Data Sources

As the above analysis sometimes raised questions about how good the data is, other data sources were collected to see if there is a dataset that performs better. This was done by appling the 'closing the water balance' method. This analysis can be found in: *DIFFERENT_DATASOURCES_ASSESSMENT.ipynb*. 

The data sources used from Earth Engine are:
- Precipitation:'NASA/GPM_L3/IMERG_V07', 'ECMWF/ERA5_LAND/DAILY_AGGR', 'UCSB-CHG/CHIRPS/DAILY'
- Evaporation:'NASA/GLDAS/V021/NOAH/G025/T3H', 'ECMWF/ERA5_LAND/DAILY_AGGR', 'MODIS/006/MOD16A2'
- Storage:'NASA/GRACE/MASS_GRIDS/MASCON_CRI' (lwe_thickness and uncertainty)

The Kalman filtering and smoothing is used to close the water balance. It can be seen that there is quite some variance between the different data sources and especially the precipitation data all falls quite often outside of the post exterior variance band. 

To see if one of the other data combinations might give a better result 6 data combinations (all datasets of which daily data was available) are plotted in the Budyko framework. 

## Conclusion

From the above analysis it can be concluded that there is no clear, definitive effect to be seen of deforestion in 1995 to be seen on the river discharge of the Cagayan de Oro river. However, this does not mean there is no effect, but several factors can be of influence:
- The data used is not reliable enough.
- Other factors that influence the water balance (e.g. precipitation) play a more important role on the water fluxes and eventually discharge in the Cagayan de Oro basin, and especially with only a ~10% decrease in forest cover the effect of deforestation can not be seen.
- The deforestation can also not have been to a completely sealed surface, but to for example cropland, which still contains quite some vegetation and thus storage and transpiration.
